```python
# access Odometer (KM) column 
car_sales["Odometer (KM)"] 
```

```python
# update Odometer to miles 
car_sales["Odometer (KM)"] = round(car_sales["Odometer (KM)"] * 0.621371) 
# rename the column to Odometer (Miles) 
car_sales.rename(columns={'Odometer (KM)': 'Odometer (Miles)'}, inplace=True) 
```

```python
car_sales
```

```python
car_sales
```

```python
# how many of them are toyota 
len(car_sales[car_sales["Make"] == "Toyota"]) 
```

```python
# add Electric column which is boolean 
car_sales["Electric"] = True 
```

```python
car_sales
```

```python
# only cars that are more than $7K are electric 
# car_sales["Electric"] = car_sales["Price"] > 7000 # not going to work since Price is string 
```

```python
car_sales["Electric"] = car_sales["Price"].replace('[\$,]','', regex=True).astype(float) > 7000 
```

```python
car_sales
```

```python
# add Is Van column 
car_sales["Is Van"] = car_sales["Doors"] > 4 
```

```python
car_sales
```

```python
car_sales.loc[car_sales["Make"] == "Toyota", "Price"] = np.nan 
```

```python
car_sales
```

```python
# replace NaN with value 
# car_sales["Price"].fillna(0.0, inplace=True) 
```

```python
# convert Price column into int  
car_sales["Price"].fillna(0, inplace=True) 

#.replace('[\$,]','', regex=True).astype(int) 
```

```python
#round(car_sales["Price"].replace('[\$,]','', regex=True).astype(float).mean()) 
```

```python
# remove $ and , 
car_sales["Price"] = car_sales["Price"].replace('[\$,]','', regex=True).astype(float).astype(int)  
```

```python
car_sales
```

```python
# calculate mean price 
mean_price = car_sales["Price"].mean().astype(int) 
```

```python
mean_price
```

```python
car_sales["Price"] = car_sales["Price"].replace(0, mean_price) 
```

```python
car_sales
```

```python
# pandas group by Make 
groups_by_make = car_sales.groupby("Make") # this will create key-value. key will be Make and Value will be groups (DataFrame)
for make, make_df in groups_by_make: 
    print(make) 
    print(make_df) 
```

```python
# get specific group 
groups_by_make.get_group("Honda") 
```

```python
groups_by_make.max() # give max for each column based on the group 
```

```python
groups_by_make.get_group("Honda")["Price"].mean().astype(int) 
```

```python
# merge 

df1 = pd.DataFrame({
    "city": ["NY", "Boston", "Denver"], 
    "temperature": [64, 42, 23]
})

df2 = pd.DataFrame({
    "city": ["Boston", "NY", "Denver"], 
    "humidity": [78, 30, 20]
})

df3 = pd.merge(df1, df2, on="city")
```
