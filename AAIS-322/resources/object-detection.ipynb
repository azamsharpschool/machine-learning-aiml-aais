```python
# Step 1: Clone YOLOv5 repository
!git clone https://github.com/ultralytics/yolov5

# Step 2: Navigate to YOLOv5 directory and install dependencies
%cd yolov5
!pip install -r requirements.txt

# Step 3: Import necessary libraries
import torch
from pathlib import Path
import cv2
import matplotlib.pyplot as plt
from IPython.display import display, Image

# Step 4: Load the YOLOv5 model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)
```

```python
from PIL import Image
```

```python
import matplotlib.pyplot as plt

image_path = '/Users/azamsharp/Desktop/machine-learning/traffic-light.png'
image = cv2.imread(image_path)

plt.imshow(image)
plt.show()

```

```python
# Step 2: Convert the image from BGR to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image_rgb)
```

```python
plt.show()
```

```python
# Step 6: Perform object detection
results = model(image_rgb)

# Step 7: Visualize the results
for _, row in results.pandas().xyxy[0].iterrows():
    cv2.rectangle(image, (int(row['xmin']), int(row['ymin'])), (int(row['xmax']), int(row['ymax'])), (255, 0, 0), 2)
    cv2.putText(image, row['name'], (int(row['xmin']), int(row['ymin'] - 10)), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10, 10))
plt.imshow(image_rgb)
plt.show()
```
